```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data_set.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# DROP UNNAMED COLUMN

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

# CHANGE WEEKLY STUDY HOURS COLUMN

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5 - 10")
df.head()
```

# GENDER DISTRIBUTION

```python
plt.figure(figsize = (6, 5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender distribution")
plt.show()
```

# From the above chart we have analyse that:
#The number of females in the data is more than the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize = (7, 5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between parent's education and student's score")
plt.show
```

#From the above chart we have concluded that the education of the parents have a good impact on their scores

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize = (8, 5))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between parent's ParentMaritalStatus and student's score")
plt.show
```

#From the above chart we have concluded that there is no/negligible impact on the student's score due to their ParentMaritalStatus
